```python
# We make a function for fetching and polishing the data a bit
def fetch_data(file_name):
    df = pd.read_csv(file_name)
    # We get rid of the extra index column as well as feature x12 since all samples have value 'True', so it doesn't give any new information
    df = df.drop(columns=['Unnamed: 0', 'x12'])
    if file_name=='TrainOnMe.csv':
        labels = df['y']
        features = df.drop(columns=['y'])
        return labels, features
    return df

Y, X = fetch_data('TrainOnMe.csv')
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.compose import ColumnTransformer

# Apply pca on the numerical columns and encode the categorical columns

preprocessor = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), ['x7']),
        ('pca', PCA(), ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x8', 'x9', 'x10', 'x11', 'x13'])])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RepeatedStratifiedKFold, cross_val_score

# After trying a few classifiers, I chose the Random Forest since it got the highest avarage accuracy from the cross-validation scores
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('random_forest', RandomForestClassifier(random_state=42))])

rskf = RepeatedStratifiedKFold(n_splits=5, n_repeats=10, random_state=42)

# Apply cross validation
scores = cross_val_score(pipeline, X_train, y_train, cv=rskf, scoring='accuracy')

print("Repeated Stratified Cross-Validation Scores:", scores)
print("Average Accuracy:", scores.mean())
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Parameter distributions we randomly search over
param_dist = {
    'random_forest__n_estimators': [100, 200, 300, 500, 700, 900, 1200],
    'random_forest__max_features': ['auto', 'log2', 'sqrt'],
    'random_forest__max_depth': [None, 10, 20, 30, 50, 70, 100, 150],
    'random_forest__min_samples_split': [2, 5, 10, 20],
    'random_forest__min_samples_leaf': [1, 2, 4, 5, 10],
    'random_forest__bootstrap': [True, False]
}

# We do a randomized search to fit the parameters for random forest, cus gridsearch takes too long time.
rand_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=150,  # Nr of different parameter combinations tested
    scoring='accuracy', 
    cv=10,  # 10-fold cross validation
    verbose=1,
    random_state=42,
    n_jobs=-1  
)

# Fit the randomized search
rand_search.fit(X_train, y_train)

print("Best Parameters:", rand_search.best_params_)
print("Best Cross-Validation Score:", rand_search.best_score_)
```

```python
# We evaluate the best model that was found using pca on numerical data, random forest and parameter fitting with random search.

from sklearn.metrics import accuracy_score

best_model = rand_search.best_estimator_
y_pred = best_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
```
